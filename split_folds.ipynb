```python
import os
import random
import json
random.seed(13)
fold_num = 10
trainval_path = './tianchiyusai'
out_json_file = './tianchi_split_folds.json'

train_list = os.path.join(trainval_path, 'ImageSets/train.txt')
train_list = open(train_list).readlines()
train_list = [x.strip() for x in train_list]

num_per_fold = len(train_list) // fold_num

train_folds = {}

for i in range(fold_num):
    cur_fold = random.sample(train_list, num_per_fold)
    train_folds[i] = cur_fold
    train_list = [x for x in train_list if x not in cur_fold]

# print(train_folds)
# for i in range(fold_num):
#     print(i, train_folds[i])
with open(out_json_file, 'w') as f:
    json.dump(train_folds, f)

with open(out_json_file) as f:
    foo = json.load(f)
    print(foo)


```
